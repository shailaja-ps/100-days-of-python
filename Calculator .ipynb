```python

def add(n1,n2):
    return n1+n2

def subtract(n1,n2):
    return n1-n2
    
def multiply(n1,n2):
    return n1*n2
    
def divide(n1,n2):
    return n1/n2

operations={ "+":add, "-":subtract, "*":multiply, "/":divide}

first_num=float(input('What is the first number?'))
while True:
    
    print( " + - * / ")
    operator=input('Pick an operation')
    next_num=float(input('What is the next number?'))
    answer=operations[operator](first_num,next_num)
    print(f"{first_num} {operator} {next_num} = {answer}")
    
    choice=input(f"Type 'y' to continue calculating with {answer} or type 'n' to end calculation")
    if choice=='y':
        first_num=answer
    else:
        break
```
