```python
alphabet = [
    'a', 'b', 'c', 'd', 'e', 'f', 'g',
    'h', 'i', 'j', 'k', 'l', 'm', 'n',
    'o', 'p', 'q', 'r', 's', 't', 'u',
    'v', 'w', 'x', 'y', 'z'
]


def encrypt(og_text,shift_number):
    encoded_text=" "
    for letter in og_text:
        if letter not in alphabet:
            encoded_text+=letter
        else:
            shifted_position=alphabet.index(letter)+shift_number
            shifted_position=shifted_position%len(alphabet)
            encoded_text+=alphabet[shifted_position]
    print(f'Here is the encoded result {encoded_text}')

    
def decrypt(og_text,shift_number):
    decoded_text=" "
    for letter in og_text:
        if letter not in alphabet:
            decoded_text+=letter
        else:
            shifted_position=alphabet.index(letter)-shift_number
            shifted_position=shifted_position%len(alphabet)
            decoded_text+=alphabet[shifted_position]
    print(f'Here is the decoded result {decoded_text}')
    

def caesar():
    if direction=='encode':
        encrypt(text,shift)
    else:
        decrypt(text,shift)
        
while True:
    direction=input("Type 'encode' to encrypt and 'decode' to descrypt :\n").lower()
    text=input('Type in your message:\n')
    shift=int(input('Type the shift number:\n'))

    caesar()

    decide=input("Do you want to continue 'yes' or 'no'").lower()
    if decide=='no':
        break
```
